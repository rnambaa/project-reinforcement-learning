```python
import pandas as pd
from utils import *
import seaborn as sns
import matplotlib.pyplot as plt

# set plot size larger
sns.set(rc={'figure.figsize':(11.7,8.27)})
# set seaborn style
sns.set_style("white")
# set seaborn color palette
sns.set_palette("Set2")
```

```python
train_data_melted = process_data('data/train.xlsx')
train_data_melted
```

```python
# plot x axis date y axis price
sns.lineplot(x='Date', y='Price', data=train_data_melted)
```

```python
# plot the data
sns.lineplot(data=train_data_melted, x='Hour', y='Price', hue=train_data_melted.Date.dt.year, linestyle='dashed', alpha=0.5)
sns.lineplot(data=train_data_melted, x='Hour', y='Price', label='Mean', color='red')

# add legend
plt.legend(title='Year')
```

```python
# plot the data
sns.lineplot(data=train_data_melted, x='Hour', y='Price', hue=train_data_melted.Date.dt.month, linestyle='dashed', alpha=0.2)
sns.lineplot(data=train_data_melted, x='Hour', y='Price', label='Mean', color='red')

# add legend
plt.legend(title='Month')
```

```python
# plot the price per day of the week, extracted from the date column
# step 1 extract the day of the week from the date column
train_data_melted['Day'] = train_data_melted['Date'].dt.day_name()
sns.lineplot(data=train_data_melted, x='Day', y='Price', hue=train_data_melted.Date.dt.year, linestyle='dashed', alpha=0.5)
sns.lineplot(data=train_data_melted, x='Day', y='Price', label='Mean', color='red')
plt.legend(title='Year')
```

```python
# plot the price per month of the year, extracted from the date column
# step 1 extract the month of the year from the date column
train_data_melted['Month'] = train_data_melted['Date'].dt.month_name()
sns.lineplot(data=train_data_melted, x='Month', y='Price', hue=train_data_melted.Date.dt.year, linestyle='dashed', alpha=0.5)
sns.lineplot(data=train_data_melted, x='Month', y='Price', label='Mean', color='red')
```

```python
# plot the price per year, extracted from the date column
# step 1: extract the year from the date column
train_data_melted['Year'] = train_data_melted['Date'].dt.year
sns.lineplot(data=train_data_melted, x='Year', y='Price', hue=train_data_melted.Date.dt.year, linestyle='dashed', alpha=0.5)
sns.lineplot(data=train_data_melted, x='Year', y='Price', label='Mean', color='red')
```

```python
# plot distributions
sns.displot(train_data_melted['Price'], kde=True)
```
